# HTML 주석 검출



```python
import requests
import re

url = 'https://dacon.io/'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    
    # HTML 주석을 찾기 위한 정규표현식 패턴
    pattern = r'<!--(.*?)-->'
    
    # re.DOTALL 옵션을 사용하면 줄바꿈 문자도 포함하여 매칭
    comments = re.findall(pattern, html, re.DOTALL)
    
    for comment in comments:
        print(comment)
else:
    print(response.status_code)
```

```python
import re

def extract_mastercard(text):
    # 방법 1: 기본 패턴 (구분자 없음)
    pattern1 = r'5[1-5]\d{14}'
    
    # 방법 2: 구분자가 있는 경우를 위해 텍스트 전처리
    # 공백이나 대시가 있는 패턴
    pattern2 = r'5[1-5](?:\d[\s-]?){14}\d'
    
    all_matches = []
    
    # 첫 번째 패턴 검색
    matches1 = re.findall(pattern1, text)
    all_matches.extend(matches1)
    
    # 두 번째 패턴 검색 후 구분자 제거
    matches2 = re.findall(pattern2, text)
    for match in matches2:
        # 구분자 제거
        cleaned = re.sub(r'[\s-]', '', match)
        # 이미 찾은 패턴과 중복 방지
        if cleaned not in all_matches:
            all_matches.append(cleaned)
    
    return all_matches

# 테스트 예제들
samples = [
    "512312번 고객 카드 번호 = 5123456789012345",
    "마스터카드: 5234-5678-9012-3456",
    "카드번호 5345 6789 0123 4567 입니다",
    "두 개의 카드: 5412345678901234 및 5523 4567 8901 2345"
]

for sample in samples:
    print(f"원본: {sample}")
    mastercard_numbers = extract_mastercard(sample)
    print(f"추출된 카드번호: {mastercard_numbers}\n")
```

```python
import re

def extract_valid_jumin(text):
    # 주민등록번호 패턴 (월/일 유효성 포함)
    # (\d{2}): 연도 두 자리
    # (0[1-9]|1[0-2]): 01~12월
    # (0[1-9]|[12][0-9]|3[01]): 01~31일
    # -[1-4]\d{6}: 하이픈 뒤 첫 자리 1-4 중 하나, 이후 6자리 숫자
    pattern = r'(\d{2})(0[1-9]|1[0-2])(0[1-9]|[12][0-9]|3[01])-[1-4]\d{6}'
    
    # 월별 일수 검증 함수
    def is_valid_date(month, day):
        month = int(month)
        day = int(day)
        
        # 월별 최대 일수 (윤년 고려 없음)
        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
        return 1 <= month <= 12 and 1 <= day <= days_in_month[month]
    
    valid_numbers = []
    matches = re.finditer(pattern, text)
    
    for match in matches:
        full_match = match.group(0)
        year = match.group(1)   # 사용하지 않지만 추출
        month = match.group(2)
        day = match.group(3)
        
        if is_valid_date(month, day):
            valid_numbers.append(full_match)
    
    return valid_numbers

# 테스트 예제
test_text = """
각종 주민번호 예시:
유효한 경우: 990101-1234567, 850505-2345678
무효한 경우: 991301-1234567 (13월), 950231-3456789 (2월 31일)
"""

result = extract_valid_jumin(test_text)
print("유효한 주민등록번호:")
for num in result:
    print(num)
```
